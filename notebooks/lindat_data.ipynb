```python
import IPython.display as ipd
import torchaudio

idx = 1
rec_dict = eval(df_transcript.iloc[idx].recognized)
audio, sr = torchaudio.load(df_transcript.iloc[idx].mp3_source)
if audio.shape[0] > 1:
    audio = audio.mean(dim=0, keepdim=True)

for s in rec_dict["segments"]:
    ipd.display(ipd.Audio(audio[:, int(s["start"] * sr): int(s["end"] * sr)], rate=sr))
    print(s["start"], s["end"])
    print(s["text"])
    print()
```
